## KNN Multi-class example

## KNN on first two components

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import datasets

iris = datasets.load_iris()
target_names = iris.target_names

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state = 1)

X_train_f = X_train[:, :2]
X_test_f = X_test[:, :2]

colors = "bry"
labels = [0,1,2]

# Plot also the training points
for i, color in zip(labels, colors):
    idx = np.where(y_train == i)
    plt.scatter(
        X_train_f[idx, 0],
        X_train_f[idx, 1],
        c=color,
        label=iris.target_names[i],
        #cmap=plt.cm.Paired,
        edgecolor="black",
        s=20,
    )
plt.legend()

neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train_f, y_train)

test_point = 15

color = ''

if(y_test[test_point] == 0):
    color = 'b'
elif(y_test[test_point] == 1):
    color = 'r'
else:
    color = 'y'

    

plt.scatter(X_test[test_point, 0], X_test[test_point, 1], c=color,marker='x')

print("Prediction: "+ iris.target_names[neigh.predict(X_test_f)[test_point]])
```

## LDA on first two components

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=2)

X_train_lda = lda.fit(np.vstack((X_train_f[:, 0], X_train_f[:, 1])).T,y_train).transform(np.vstack((X_train_f[:, 0], X_train_f[:, 1])).T)

plt.figure()

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(
        X_train_lda[y_train == i, 0], X_train_lda[y_train == i, 1], color=color, label=target_name, edgecolor="black", s=20
    )
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("LDA of IRIS dataset")

plt.show()
```

## KNN on last two components

```python
X_train_l = X_train[:, 2:]
X_test_l = X_test[:, 2:]


colors = "bry"
labels = [0,1,2]

# Plot also the training points
for i, color in zip(labels, colors):
    idx = np.where(y_train == i)
    plt.scatter(
        X_train_l[idx, 0],
        X_train_l[idx, 1],
        c=color,
        label=iris.target_names[i],
        #cmap=plt.cm.Paired,
        edgecolor="black",
        s=20,
    )
plt.legend()

neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train_l, y_train)

test_point = 15

color = ''

if(y_test[test_point] == 0):
    color = 'b'
elif(y_test[test_point] == 1):
    color = 'r'
else:
    color = 'y'

    

plt.scatter(X_test_l[test_point, 0], X_test_l[test_point, 1], c=color,marker='x')

print("Prediction: "+iris.target_names[neigh.predict(X_test_l)[test_point]])
```

## LDA on last two components

```python
lda = LinearDiscriminantAnalysis(n_components=2)

X_train_lda = lda.fit(np.vstack((X_train_l[:, 0], X_train_l[:, 1])).T,y_train).transform(np.vstack((X_train_l[:, 0], X_train_l[:, 1])).T)

plt.figure()

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(
        X_train_lda[y_train == i, 0], X_train_lda[y_train == i, 1], color=color, label=target_name, edgecolor="black", s=20
    )
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("LDA of IRIS dataset")

plt.show()
```
