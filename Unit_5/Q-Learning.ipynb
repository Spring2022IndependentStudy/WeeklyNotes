# An introduction to Q-Learning

## The original formulation of Q-Learning

Q-Learning is one of the more basic reinforcement learning algorithms because of its "model-free reinforcement learning" nature. Essentially, Q-learning does not use a transition probability distribution of the environment and only seeks to model actions that it can perform. In contrast to a model-based approach, a model-free algorithm has the agent learn policies directly. HIstorically, Q-Learning was one of the first major breakthroughts in reinfocement learning as it was the first algorithm to provide researchers with a convergence guarantee. Q-Learning, like many other algorithms, has both pros and cons. Q-Learning considers future steps by using previously learned "states" that have been investigated, and stores this information in a "Q-Table." For every action done by a state, a positive or negative reward must be included in the policy table, Q table. 

The model begins with a fixed epsilon value that reflects movement randomness. The model starts with a fixed epsilon value, which represents the randomization of movements. Over time, the randomization decreased based upon the epsilon decay value. Furthermore, if the current state of the agent is considered to be new or unexplored, the agent will just produce a randomly generated move, in an attempt to better learn the environment.

This form of learning is great when there are a limited number of moves or the environment is not complicated, as the agent remembers past moves and repeats them with ease.
However, for more complex environments with a significantly larger number of states, the Q-Table will fill up rapidly, causing large training times. The issue is, Q-Learning does not predict but rather deals in absolutes the majority of the time: the state has already been considered and the action to produce is already known, or the state is unknown, and a random action has to be applied upon.

## A transition to Deep Q-Learning

As previously mentioned above, a weakness of classic Q-Learning is that it doesn't scale well in environment that are complex due to the high cost of storing all the calculated values of Q along with the ever expanding cost to traverse through the Q-table to calculate the next set of Q-values. However, with neural networks, there is an optmization that we can make to lessen the impact of this problem.

We must dedicate two neural networks to solve $y_i$ and $Q(s,a)$ which are called the target and the Q-network respectively. The weights and biases of the target-network correspond with the parameter of the Q-network at an earlier point of time.

An optimization that we can apply to stabalize the Deep Q-model is the use of experience replay. Experience replay is essentially the act of storing $< s, s, a, r >$ where
>s=state of the agent
>a'=action taken in the state s by the agent
>r=immediate reward received in state s for action a'
>s'=next state of the agent after state s

We take random batches from the experience replay memory to calculate the TD-Target and this regularizes our model and allows us to learn from previous experience multiple times.
